```python
import pandas as pd
```

```python
#files to load
school_data_to_load = "School_District_Analysis/Resources/schools_complete.csv"
student_data_to_load = "School_District_Analysis/Resources/students_complete.csv"
```

```python
# Read the school data file and store it in a Pandas DataFrame.
school_data_df = pd.read_csv(school_data_to_load)
school_data_df
```

```python
#Read the student data file and sttore it in Pandas DataFrame.
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.head()
```

```python
#Determine if there are any missing values in the student data.
student_data_df.count()
```

```python
#Determine data types for the school DataFrame.
school_data_df.dtypes
```

```python
#Add each prefi and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ", "Ms. ", "Mrs. ", "Miss "," MD"," DDS", " DVM", " PhD"]
```

```python
#Iterate through the words in "prefixes_suffixes" list and replace them with an empty space, "".
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word, "", regex=True)
```

```python
student_data_df
```

```python
#Combine the data into a single dataset.
school_data_complete_df = pd.merge(student_data_df, school_data_df, on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
#Get the total number of students.
student_count = school_data_complete_df.count()
student_count
```

```python
#Calculate the total budget on the school data frame, not the completed school data frame.
total_budget = float(school_data_df["budget"].sum())
total_budget
```

```python
#Calculate the average reading score.
average_reading_score = float(school_data_complete_df["reading_score"].mean())
average_reading_score
```

```python
#Calculate the average math score.
average_math_score = float(school_data_complete_df["math_score"].mean())
average_math_score
```

```python
#Determining the passing scores for math and reading.
passing_math = school_data_complete_df["math_score"] >= 70
passing_reading = school_data_complete_df["reading_score"] >= 70
```

```python
passing_math = school_data_complete_df[school_data_complete_df["math_score"] >= 70]
passing_math.head()
```

```python
#Putting all the students that with passing reading scores in a new DataFrame.
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]
passing_reading
```

```python
#Calculate the number of students passing math.
passing_math_count = passing_math["student_name"].count()
passing_math_count
```

```python
student_count
```

```python
student_real_count = school_data_complete_df["Student ID"].count()
student_real_count
```

```python
#Calculate the percent that passed math.
passing_math_percentage = passing_math_count / float(student_real_count) * 100
passing_math_percentage
```

```python
#Calculate the percent that passed reading.
passing_reading_percentage = passing_reading_count / float(student_real_count) * 100
passing_reading_percentage
```

```python
#Calculate the students who passed both math and reading.
passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)]

passing_math_reading.head()
```

```python
#Calculate the number of students who passed both math and reading.
overall_passing_math_reading_count = passing_math_reading["student_name"].count()
overall_passing_math_reading_count
```

```python
overall_passing_percentage = overall_passing_math_reading_count / student_real_count * 100
overall_passing_percentage
```

```python
district_summary_df = pd.DataFrame(
    [{"Total Schools": len(school_count_2),
      "Total Students": student_real_count,
      "Total Budget": total_budget,
      "Average Math Score": average_math_score,
      "Average Reading Score": average_reading_score,
      "% Passing Math": passing_math_percentage,
      "% Passing Reading": passing_reading_percentage,
      "% Overall Passing": overall_passing_percentage}])
district_summary_df
```

```python
#Format the "Total Students" to have the comma for a thousands separator.
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format )
district_summary_df["Total Students"]
```

```python
#Formating "Total Budget".
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format )
```

```python
district_summary_df["Total Budget"]
```

```python
# Format the remaining columns.
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)

district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)

district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.0f}".format)

district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.0f}".format)

district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.0f}".format)
```

```python
#Display the DataFrame
district_summary_df
```

```python
#Reorder the columns in the order that you want them to appear.
new_column_order = ["Total Schools", "Total Students", "Total Budget","Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]

#Assign disttrict summary df the new column order.
district_summary_df = district_summary_df[new_column_order]
district_summary_df
```

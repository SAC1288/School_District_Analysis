```python
import pandas as pd
```

```python
#files to load
school_data_to_load = "School_District_Analysis/Resources/schools_complete.csv"
student_data_to_load = "School_District_Analysis/Resources/students_complete.csv"
```

```python
# Read the school data file and store it in a Pandas DataFrame.
school_data_df = pd.read_csv(school_data_to_load)
school_data_df
```

```python
#Read the student data file and sttore it in Pandas DataFrame.
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.head()
```

```python
#Determine if there are any missing values in the school data.
school_data_df.count()
```

```python
#Determine if there are any missing values in the student data.
student_data_df.count()
```

```python
#Determine if tthere are any missing values in the school data.
school_data_df.isnull()
```

```python
#Determine if there are any missing values in the student data.
student_data_df.isnull()
```

```python
#Another method for determining if there are missing values in the student data.
student_data_df.isnull().sum()
```

```python
#Using notnull function to determine if there are missing values within the school data.
school_data_df.notnull()
```

```python
#Using the notnull function on the student data series.
student_data_df.notnull().sum()
```

```python
#Determine data types for the school DataFrame.
school_data_df.dtypes
```

```python
#Add each prefi and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ", "Ms. ", "Mrs. ", "Miss "," MD"," DDS", " DVM", " PhD"]
```

```python
#Iterate through the words in "prefixes_suffixes" list and replace them with an empty space, "".
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word, "", regex=True)
```
